**Task_1** Load Libraries and Data

```python
# Import the numpy and pandas libraries.
import numpy as np
import pandas as pd

# Read the data file into a DataFrame named df.
df = pd.read_csv('/workspaces/Data-Cleaning-Project/data/raw data/Cafe Sales.csv')
```

Task_3 Check for Duplicates

```python
# Check for any complete duplicate rows.
print(f' Number Of Duplicate Record : {sum(df.duplicated())}')
```

**Task_4** Identify Bad and Missing Data

```python
# Calculating  the percentage of missing values for each column.
df.isna().sum()*100/df.shape[0]
```

```python
# Identifying the unique values in the categorical columns.
df_column = set(df.columns)
exclude = {'Transiaction ID'}
df_column = df_column.difference(exclude)
for column in df_column:
   print(f'{column}: \n{df[column].unique()}', end='\n\n')
```

**Task_5** Clean Bad Values

```python
# Replacing all instances of 'ERROR' and 'UNKNOWN' across the entire DataFrame.
df = df.replace(['ERROR', 'UNKNOWN'], np.nan)
```

```python
# Percentage of missing values after replacing by NaN.
df.isna().sum()*100/df.shape[0]
```

**Task_6** Correct Data Type

```python
# Converting the 'Quantity' column to the 'Int64' type.
df['Quantity'] = df['Quantity'].astype('Int64')

# Converting the 'Price Per Unit' and 'Total Spent' columns to the 'float' type.
df[['Price Per Unit', 'Total Spent']] = df[['Price Per Unit', 'Total Spent']].astype(float)

# Converting the 'Transaction Date' column to 'datetime' type.
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df.info()
```

**Task_7** Handle Missing Data

```python
# Drop any rows where 'Transaction Date' is missing.
df = df.dropna(subset= 'Transaction Date')
df.info()
```

```python
# Fill missing values in 'Payment Method' and 'Location' with the placeholder string 'INA'.
df[['Payment Method', 'Location']] = df[['Payment Method', 'Location']].fillna('INA')
df.info()
```

```python
# Fill missing 'Item' values with the mode.
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
df.info()
```

```python
df.describe()
```

```python
# Fill missing 'Quantity' values with the median.
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df.info()
```

```python
df.groupby('Item')['Price Per Unit'].value_counts()
```

```python
# Filling NaNs with the mode of each item's price.
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(lambda x : x.fillna(x.mode()[0]))
df.info()
```

```python
# Filling 'Total Spent' values by calculating (Quantity * Price Per Unit)
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
df.info()
```

Task_8 Save the Cleaned Dataset

```python
# Save the final, cleaned DataFrame to a new CSV file.
df. to_csv('/workspaces/Data-Cleaning-Project/data/raw data/processed/Cafe Sales (clened).csv')
```
